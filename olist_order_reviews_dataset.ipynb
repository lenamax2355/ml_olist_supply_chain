```python
import pandas as pd
import datetime as dt
ore=pd.read_csv('olist_order_reviews_dataset.csv')
ore.head()
```

```python
ore.isna().sum()
```

```python
ore.describe(include='all')
```

```python
ore.dtypes
```

```python
# 1st we have to cover object dataypes to datetime format so we can get date, year, etc information from it easily
ore[['review_creation_date','review_answer_timestamp']] = ore[['review_creation_date','review_answer_timestamp']].apply(pd.to_datetime)
```

```python
ore.dtypes
```

```python
# we can clear see that review_creation_date doesnot contain hour,min and  sec information so we can get only year,month and date info
ore['review_creation_date_yr'] = ore['review_creation_date'].dt.year
ore['review_creation_date_mo'] = ore['review_creation_date'].dt.month
ore['review_creation_date_dy'] = ore['review_creation_date'].dt.day
```

```python
ore['review_answer_timestamp_yr'] = ore['review_answer_timestamp'].dt.year
ore['review_answer_timestamp_mo'] = ore['review_answer_timestamp'].dt.month
ore['review_answer_timestamp_dy'] = ore['review_answer_timestamp'].dt.day
ore['review_answer_timestamp_hr'] = ore['review_answer_timestamp'].dt.hour
ore['review_answer_timestamp_mi'] = ore['review_answer_timestamp'].dt.minute
ore['review_answer_timestamp_sec'] = ore['review_answer_timestamp'].dt.second
```

```python
ore
```

```python
ore=ore.drop(['review_creation_date','review_answer_timestamp'],axis=1)
```

```python
ore.isna().sum()
```

```python
ore.to_csv('new_olist_order_reviews_dataset.csv',index=False)
```
